### Exploratory Analysis on Dataset

```python
# reading the churn data
df = pd.read_csv('Telco-Customer-Churn.csv')
df.info()
```

```python
# Total Charges column needs to be converted to numeric as it seems like a numeric column

df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
# we can see most of the columns are of type object and on closer look they are all categorical datatype. Let us visualize 
# first and then convert the columns to dummy integer codes.

def category_plot(column):
    plot = df.groupby([column,'Churn']).size().reset_index().pivot(columns='Churn', index=column, values=0)
    plot.plot(kind='bar',stacked=True);
```

```python
category_plot('gender') 
# Gender does not seem to have an effect on the churn
```

```python
category_plot('Partner') 
# The customer has partner or not, does seem to have some effect on the churn
```

```python
category_plot('Dependents') 
# Customers who have no dependents are more likely to churn than customers who have dependents
```

```python
category_plot('PhoneService')
# There are not many customers did not sign up for phone service, whether customer have phone service or not does not seem to have an effect on the churn
```

```python
category_plot('MultipleLines')
# Whether customer signed up for MultipleLines or not does not seem to have an effect on the churn
```

```python
category_plot('InternetService')
# Customers who signed up for Fiber optic are most likely to churn, almost 50% of them churned
```

```python
category_plot('OnlineSecurity')
# Customers who did not sign up for OnlineSecurity are most likely to churn
```

```python
category_plot('OnlineBackup')
# Customers who did not sign up for OnlineBackUp are most likely to churn
```

```python
category_plot('DeviceProtection')
#Customers who did not sign up for DeviceProtection are most likely to churn
```

```python
category_plot('TechSupport')
# Customers who did not sign up for TechSupport are most likely to churn
```

```python
category_plot('StreamingTV')
```

```python
category_plot('StreamingMovies')
# From above two plots, we can see that customers without internet service have a very low churn rate
```

```python
category_plot('Contract')
# It is obvious that contract term does have an effect on churn. 
# There were very few churns when customers have a two-year contract. And most churns occurred on customers with a month-to-month contract
```

```python
category_plot('PaperlessBilling')
```

```python
category_plot('PaymentMethod')
# PaymentMethod does seem to have an effect on churn, in particular, pay by electronic check has the highest percentage churning rate
```

```python
category_plot('SeniorCitizen')
# There are not many senior citizens in the data. It seems whether customers are seniors citizens or not does not have an effect on the churning rate.
```

```python
# Exploring numeric features using seaborn kernel-density-estimation
sns.kdeplot(df['tenure'].loc[df['Churn'] == 'No'], label='not churn', shade=True);
sns.kdeplot(df['tenure'].loc[df['Churn'] == 'Yes'], label='churn', shade=True);
# Not churned customers have a much longer average tenure than the churned customers.
```

```python
sns.kdeplot(df['MonthlyCharges'].loc[df['Churn'] == 'No'], label='not churn', shade=True);
sns.kdeplot(df['MonthlyCharges'].loc[df['Churn'] == 'Yes'], label='churn', shade=True);
# Churned customers paid higher average monthly fee than not-churned customers.
```
